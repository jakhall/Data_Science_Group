```python
#load in the training and test data
Train = pd.read_csv('training_data_mbti_one_comment.csv')
Test = pd.read_csv('testing_data_mbti_one_comment.csv')
```

```python
#splitting test and train data into features needed
train_data = Train.loc[:,'abandon':'zone']
train_target_1 = Train.loc[:,'t1']
train_target_2 = train_target_1.replace(['I'],0)
train_target = train_target_2.replace(['E'],1)

#test
test_data = Test.loc[:,'abandon':'zone']
test_target_1 = Test.loc[:,'t1']
test_target_2 = test_target_1.replace(['I'],0)
test_target = test_target_2.replace(['E'],1)

test_target2 = 
test_target3 = 
test_target4 = 

test_target2
#train_target

#reshaping data
train_target = train_target.values.reshape(-1,1)
test_target = test_target.values.reshape(-1,1)

#print(train_data.head)
#print(test_data.head)
```

```python
#creating the svm
from sklearn.svm import LinearSVR
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

#Hyperparameters to test
parameters = [{'kernel': ['rbf'],
               'gamma': [0.2, 0.5],
                'C': [1, 10]},
              {'kernel': ['linear'], 'C': [1, 10]}]

#When ready tune the hyperparameters
svm_class = GridSearchCV(svm.SVC(decision_function_shape='ovr'), parameters, cv=3)
svm_class.fit(train_data, train_target)
print(svm_class.best_params_)

# fit a linear support vector machine regressor for testing
#epsilon_svm = 1

#svm_class = LinearSVR(epsilon_svm)
#svm_class.fit(train_data, train_target)

predictions = svm_class.predict(test_data)


#confusion matrix
cm = metrics.confusion_matrix(test_target, predictions)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
#creating a clasification report  
print(classification_report(test_target,predictions)) 
#main error is always predicts 0.
```
